```python
import pandas as pd
```

```python
df = pd.read_csv('data/data.csv')
```

```python
df.head()
```

```python
new_list = []
for price in df["price"]:
    new_list.append(price)
```

```python
def count_saving_year(x):
    year_amount = []
    for price in x:
        calc = (price * 0.3) / (40_000 * 0.781 - 15_900)
        year_amount.append(round(calc/12, 1))
    return year_amount
```

```python
test_year_count = count_saving_year(new_list)
```

```python
year_column = pd.Series(test_year_count)
```

```python
year_column
```

```python
df["year_column"] = year_column
```
